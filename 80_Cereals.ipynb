```python
plt.figure(figsize=(10,6))
plt.title("Overall cereals data distinution")
sns.histplot(data=df)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['mfr'],bins=20, kde=True, color='red')
plt.title("Distribution of manufacturer")
plt.xlabel('mfr')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['type'],bins=20, kde=True, color='blue')
plt.title('Distribution of Cold or Hot Types of cereal')
plt.xlabel('type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['protein'],bins=20, kde=True, color='orange')
plt.title('Distribution of protein')
plt.xlabel('protein')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['fat'],bins=20, kde=True, color='purple')
plt.title('Distribution of fat')
plt.xlabel('fat')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['sodium'],bins=20, kde=True, color='pink')
plt.title('Distribution of sodium')
plt.xlabel('sodium')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['fiber'],bins=20, kde=True, color='yellow')
plt.title('Distribution of fiber')
plt.xlabel('fiber')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['rating'],bins=20, kde=True, color='black')
plt.title('Distribution of Cereal Ratings')
plt.xlabel('Rating')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='calories', y='protein')
plt.title('Calories vs Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='calories', y='rating')
plt.title('Calories vs Rating')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.show()
```

```python
cereal_sorted = df.sort_values(by='calories', ascending=False)
cereal_sorted.head()
```

```python
plt.figure(figsize=(15, 7))
plt.title("Brands and Their Ratings")
plt.xticks(rotation=90)
sns.barplot(data=cereal_sorted, x=cereal_sorted['name'], y=cereal_sorted['rating'])
```

```python
plt.figure(figsize=(10,6))
plt.title("Relationship between sugar content and Rating")
plt.xticks(rotation=90)
sns.regplot(data=df, x=df['sugars'], y=df['rating'])
```

```python
plt.figure(figsize=(10,6))
plt.title("Relationship between fat content and Rating")
plt.xticks(rotation=90)
sns.regplot(data=df, x=df['fat'], y=df['rating'])
```
